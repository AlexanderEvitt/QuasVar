```python
import numpy as np
from pyESN import ESN
from matplotlib import pyplot as plt
import lightkurve as lk
import pickle
from tqdm import tqdm
import os
import math
import pandas as pd

r = 78
esn = ESN(2, 1,
              n_reservoir=550,
              spectral_radius=0.95,
              sparsity=0.5,
              noise=7,
              input_shift=None,
              input_scaling=None,
              teacher_forcing=True,
              feedback_scaling=None,
              teacher_scaling=None,
              teacher_shift=None,
              random_state=0)

predictions = []
true_masses = []
masses =  pd.read_csv('F:/catalogues/final_quasar_catalogue.csv', sep=',', usecols=['MBH_MgII'],header=0,squeeze=True)
mags =  pd.read_csv('F:/catalogues/final_quasar_catalogue.csv', sep=',', usecols=['GMAG'],header=0,squeeze=True)
for file in tqdm(os.listdir('F:/sector_lightcurves/')[:1000]):
    row = int(file.split('-')[0])
    if math.isnan(masses[row]):
        continue
    with open('F:/sector_lightcurves/'+file,'rb') as rfile:
        lc = pickle.load(rfile)
    lc.quality = np.zeros(len(lc.flux),dtype='int')
    lc = lc.remove_nans()
    mass = masses[row]
    mag = mags[row]

    trainlen = 600
    input_data = np.vstack((lc.flux[:trainlen],mag*np.ones(trainlen))).transpose()
    pred_training = esn.fit(input_data,mass*np.ones(trainlen))
    prediction = esn.predict(input_data)
    predictions.append(np.mean(prediction))
    true_masses.append(mass)
print(np.sqrt(np.mean((np.array(predictions)-np.array(true_masses))**2)))
```

```python
n_reservoir= 500
sparsity   = 0.2
rand_seed  = 23

radius_set = np.linspace(1,3,10)
noise_set = np.linspace(0,2,10)

radius_set_size  = len(radius_set)
noise_set_size = len(noise_set)

trainlen = 600

loss = np.zeros([radius_set_size, noise_set_size])

for l in range(radius_set_size):
    rho = radius_set[l]
    for j in range(noise_set_size):
        noise = noise_set[j]


        esn = ESN(n_inputs = 1,
          n_outputs = 1, 
          n_reservoir = n_reservoir,
          sparsity=sparsity,
          random_state=rand_seed,
          spectral_radius = rho,
          noise=noise)
        
        errors = []
        for file in os.listdir('F:/sector_lightcurves/')[:10]:
            row = int(file.split('-')[0])
            if math.isnan(masses[row]):
                continue
            with open('F:/sector_lightcurves/'+file,'rb') as rfile:
                lc = pickle.load(rfile)
            lc.quality = np.zeros(len(lc.flux),dtype='int')
            lc = lc.remove_nans()
            mass = masses[row]

            trainlen = 600
            pred_training = esn.fit(lc.flux[:trainlen],mass*np.ones(trainlen))
            prediction = esn.predict(lc.flux[:trainlen])
            errors.append(np.mean(prediction)-mass)
        loss[l, j] = np.sqrt(np.mean(np.array(errors)**2))       
        #print('rho = ', radius_set[l], ', noise = ', noise_set[j], ', MSE = ', loss[l][j] )

plt.imshow(loss)
plt.xlabel('Noise')
plt.ylabel('Spectral Radius')
plt.colorbar()
print("Noise values: {}".format(noise_set))
print("Radius values: {}".format(radius_set))
```

```python
masses =  pd.read_csv('F:/catalogues/final_quasar_catalogue.csv', sep=',', usecols=['MBH_MgII'],header=0,squeeze=True)
mags =  pd.read_csv('F:/catalogues/final_quasar_catalogue.csv', sep=',', usecols=['GMAG'],header=0,squeeze=True)

var_set = np.linspace(0,10,100,dtype='int')
var_set_size  = len(var_set)

trainlen = 600

loss = []

for v in tqdm(range(var_set_size)):
    var = var_set[v]

    esn = ESN(2, 1,
              n_reservoir=550,
              spectral_radius=0.95,
              sparsity=0.5,
              noise=var,
              input_shift=None,
              input_scaling=None,
              teacher_forcing=True,
              feedback_scaling=None,
              teacher_scaling=None,
              teacher_shift=None,
              random_state=0)

    errors = []
    for file in os.listdir('F:/sector_lightcurves/')[:10]:
        row = int(file.split('-')[0])
        if math.isnan(masses[row]):
            continue
        with open('F:/sector_lightcurves/'+file,'rb') as rfile:
            lc = pickle.load(rfile)
        lc.quality = np.zeros(len(lc.flux),dtype='int')
        lc = lc.remove_nans()
        mass = masses[row]
        mag = mags[row]
        input_data = np.vstack((lc.flux[:trainlen],mag*np.ones(trainlen))).transpose()
        pred_training = esn.fit(input_data,mass*np.ones(trainlen))
        prediction = esn.predict(input_data)
        errors.append(np.mean(prediction)-mass)
    loss.append(np.sqrt(np.mean(np.array(errors)**2)))    

plt.plot(var_set,loss)
plt.title("Loss over a certain variable")
print(loss.index(min(loss)),min(loss))
```
