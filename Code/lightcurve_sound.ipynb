```python
import numpy as np
from scipy.io.wavfile import write
from scipy.interpolate import interp1d
from IPython.display import Audio
import pickle
import lightkurve
import matplotlib.pyplot as plt
```

```python
file = 'blazar.pickle'
sigma = 10
amount = 3

with open(file,'rb') as rfile:
    lc = pickle.load(rfile)
lc = lc.remove_nans().remove_outliers(sigma=sigma)

data = lc.flux * 10000
#Interpolate
for _ in range(amount):
    new_data = []
    for n in range(len(data)-1):
        y = (data[n] + data[n+1])/2
        new_data.append(data[n])
        new_data.append(y)
    data = new_data

print(len(data))
plt.plot(data,color='blue')
plt.title('TIC ' + str(lc.targetid))
Audio(data, rate=2000)
```

```python
def audio_from_lc(file,median=30,seconds=30,frequency=440,rate=44100):
    with open(file,'rb') as rfile:
        lc = pickle.load(rfile).remove_nans()
    fluxes = lc.flux + median
    length = int((seconds*rate)/len(fluxes))

    fluxes = lc.flux + median

    volumes = np.sin((2*np.pi*np.arange(length*len(fluxes))/(rate/frequency)))
    for n in range(len(fluxes)):
        volumes[length*n:length*(n+1)] = volumes[length*n:length*(n+1)] * fluxes[n]

    plt.plot(volumes,color='blue')
    plt.ylim(-5*np.std(volumes),5*np.std(volumes))
    plt.title('TIC ' + str(lc.targetid))
    return Audio(volumes, rate=rate)

def multi_freq_audio_from_lc(file,median=30,seconds=30,frequencies=[440],ratios=[1],rate=44100):
    volumes_list = []
    for n in range(len(frequencies)):
        with open(file,'rb') as rfile:
            lc = pickle.load(rfile).remove_nans()
        fluxes = lc.flux + median
        length = int((seconds*rate)/len(fluxes))

        fluxes = ratios[n]*(lc.flux + median)

        volumes = np.sin((2*np.pi*np.arange(length*len(fluxes))/(rate/frequencies[n])))
        for n in range(len(fluxes)):
            volumes[length*n:length*(n+1)] = volumes[length*n:length*(n+1)] * fluxes[n]
        volumes_list.append(volumes)
    volumes = np.average(np.array(volumes_list),axis=0)
   
    plt.plot(volumes,color='blue')
    plt.ylim(-5*np.std(volumes),5*np.std(volumes))
    plt.title('TIC ' + str(lc.targetid))
       
    return Audio(volumes, rate=rate)

def ratios_from_spectra(file,maximum=440,downscale=10):
    with open(file,'rb') as rfile:
        sp = pickle.load(rfile)
    print("There are {0} data points present in the spectra file. After downsizing this will result in a {1}x[Length of Light Curve] dimensional volumes array.".format(len(sp[0]),int(len(sp[0])/downscale)))
    frequencies = maximum*((sp[0]-min(sp[0]))/(max(sp[0])-min(sp[0])))
    frequencies = [frequencies[i] for i in range(len(frequencies)) if i % downscale == 0]
    ratios = ((sp[1]-min(sp[1]))/(max(sp[1])-min(sp[1])))
    ratios = [ratios[i] for i in range(len(ratios)) if i % downscale == 0]
    return frequencies,ratios

#frequencies,ratios = ratios_from_spectra('F:/sdss_database/spectra/3sp.pickle',downscale=10)
audio_from_lc('blazar.pickle',median=5,seconds=10,frequency=330)
```

```python
file='F:/sdss_database/spectra/10sp.pickle'
downscale=5
rang=110
minimum=330

with open(file,'rb') as rfile:
    sp = pickle.load(rfile)
print("There are {0} data points present in the spectra file. After downsizing this will result in a {1}x[Length of Light Curve] dimensional volumes array.".format(len(sp[0]),int(len(sp[0])/downscale)))
frequencies = rang*((sp[0]-min(sp[0]))/(max(sp[0])-min(sp[0])))+minimum
frequencies = [frequencies[i] for i in range(len(frequencies)) if i % downscale == 0]
ratios = ((sp[1]-min(sp[1]))/(max(sp[1])-min(sp[1])))
ratios = [ratios[i] for i in range(len(ratios)) if i % downscale == 0]

file='F:/sdss_database/lightcurves/10lc.pickle'
median=20
seconds=10
rate=44100

volumes_list = []

with open(file,'rb') as rfile:
    lc = pickle.load(rfile).remove_nans().remove_outliers(10)
fluxes = lc.flux+median
length = int((seconds*rate)/len(fluxes))

for n in range(len(frequencies)):
    volumes = ratios[n]*np.sin((2*np.pi*np.arange(length*len(fluxes))/(rate/frequencies[n])))
    for n in range(len(fluxes)):
        volumes[length*n:length*(n+1)] = volumes[length*n:length*(n+1)] * fluxes[n]
    volumes_list.append(volumes)
volumes = np.average(np.array(volumes_list),axis=0)

plt.plot(volumes,color='blue')
plt.ylim(-5*np.std(volumes),5*np.std(volumes))
plt.title('TIC ' + str(lc.targetid))

Audio(volumes, rate=rate)
```
