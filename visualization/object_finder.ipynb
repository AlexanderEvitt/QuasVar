```python
import pickle
import lightkurve as lk
from matplotlib import pyplot as plt
from os import listdir
import time
import numpy as np
```

```python
for file in listdir('E:/lightcurves/star/full/'):

    with open('E:/lightcurves/star/full/'+file,'rb') as rfile:
        lc = pickle.load(rfile)

    lc = lc.remove_nans().remove_outliers()

    lc = lc.flatten()
    plt.figure()
    period = np.linspace(1, 20, 10000)
    bls = lc.to_periodogram(method='bls', period = period, frequency_factor=500)
    bls.plot(title=file)
    plt.savefig('E:/vis/planets/' + file[:-7] + '.jpg')
    plt.close()
```
